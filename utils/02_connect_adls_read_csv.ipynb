```python
# Cell 1 — Define storage path (no authentication code needed!)
# Unity Catalog handles auth automatically via our External Location

base_path = "abfss://adventureworks@adwsadatabricksdev.dfs.core.windows.net"

# List all files to confirm access
files = dbutils.fs.ls(base_path)
print(f"✅ Connected! Files found: {len(files)}\n")
for f in files:
    size_mb = round(f.size / (1024*1024), 2)
    print(f"   📄 {f.name:<45} {size_mb} MB")
```

```python
# Cell 2 — Read SalesTerritory into a PySpark DataFrame
# Small file = perfect for testing our first read

df_territory = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(f"{base_path}/salesterritory.csv")

print(f"✅ Rows loaded : {df_territory.count()}")
print(f"   Columns     : {len(df_territory.columns)}")
print(f"\n📋 Schema:")
df_territory.printSchema()
```

```python
# Cell 3 — Read the SalesTerritory CSV (small file, great for testing)
# inferSchema=True tells Spark to automatically detect column data types
# header=True tells Spark the first row contains column names

df_territory = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(f"{base_path}/salesterritory.csv")

# How many rows did we get?
print(f"✅ Rows loaded: {df_territory.count()}")
print(f"   Columns    : {len(df_territory.columns)}")
print(f"\n📋 Schema (column names + data types):")
df_territory.printSchema()
```

```python
# Cell 3 — Preview the data
# SQL equivalent: SELECT * FROM SalesTerritory

df_territory.show(10, truncate=False)
```
